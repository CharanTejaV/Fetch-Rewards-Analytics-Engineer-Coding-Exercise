```python
###Import all the required libraries
import pandas as pd
from pandas import json_normalize
import json
from datetime import datetime

json_file_path = 'C:/Users/princ/PycharmProjects/pythonProject/Assessment/users.json'

# Read JSON file line by line into a list
with open(json_file_path, 'r') as file:
    lines = file.readlines()

# Load each line as a separate JSON object
data = [json.loads(line) for line in lines]

# json_normalize to flatten the nested structure
users = json_normalize(data)
users.head(10)
```

```python
users.rename(columns={
    '_id.$oid': 'user_id',
    'createdDate.$date':'created_date',
    'lastLogin.$date':'lastlogin_data'}, inplace=True)
users.info()
```

```python
users['created_date'] = users['created_date'].dropna().apply(lambda x: datetime.fromtimestamp(x / 1000.0))
users['lastlogin_data'] = users['lastlogin_data'].dropna().apply(lambda x: datetime.fromtimestamp(x / 1000.0))
```

```python
excel_file_path = 'C:/Users/princ/PycharmProjects/pythonProject/Assessment/users.xlsx'
users.to_excel(excel_file_path, index=False)
users.head(10)
```
