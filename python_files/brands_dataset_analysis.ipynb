```python
###Import all the required libraries
import pandas as pd
from pandas import json_normalize
import json
```

```python
json_file_path = 'C:/Users/princ/PycharmProjects/pythonProject/Assessment/brands.json'

# Reading JSON data line by line into a list
with open(json_file_path, 'r') as file:
    lines = file.readlines()
    
data = [json.loads(line) for line in lines] # Loading each line as a separate JSON object

# Use json_normalize to flatten the nested structure
brands = json_normalize(data)
brands.head(10)
```

```python
brands.info()
```

```python
""" Data Cleansing """
brands["barcode"] = brands["barcode"].astype('int64')
brands.rename(columns={
    '_id.$oid': 'id',
    'cpg.$id.$oid': 'cpg_id',
    'cpg.$ref': 'cpg_ref'
}, inplace=True)
```

```python
"""Converting JSON file into Excel file for having a structure for the data"""
excel_file_path = 'C:/Users/princ/PycharmProjects/pythonProject/Assessment/brands.xlsx'
brands.to_excel(excel_file_path, index=False)
brands.head(10)
```

```python
"""Quantifying missing data"""
print(brands.isna().sum())
```

```python
percentage_null_values = brands.isna().mean()
print("Percentage of null values")
for key, value in percentage_null_values.items():
  if value>0:
    print(f"{key}: {round(value * 100, 2)}%")
```

TopBrand: A large number of nulls (more than half) could mean missing opportunities to highlight top brands or inconsistencies in how brands are promoted.
Category Code: It is crucial for integrating with external datasets or performing detailed category-level analysis.
BrandCode: Missing brand codes could complicate the process of matching records and analyzing each brand performance.
Category: 13% null values might make it difficult to understand the brand performance within specific market segments.

```python
"""Checking duplicates"""
duplicates = brands[brands.duplicated()]
if not duplicates.empty:
    print('Duplicate rows (except for their first occurrence):')
    print(duplicates)
else:
    print('No duplicate rows found.')
```
